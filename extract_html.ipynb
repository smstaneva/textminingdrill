1. Extract html from a Webpage

```python
from urllib import request
```

```python
from bs4 import BeautifulSoup
```

```python
webpage = request.urlopen("http://www.michaelpoliza.com/en").read()
```

```python
webpage
```

```python
page = BeautifulSoup(webpage, 'html.parser')
```

```python
page
```
